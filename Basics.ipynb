```python
import csv 
with open("guns.csv","r") as f:
    reader=csv.reader(f)
    data=list(reader)
    print(data[:5])
```

```python
header=data[:1]
data=data[1:]
print(header)
print(data[:5])
```

```python
years=[row[1] for row in data[:1]]
year_counts={}
for year in years:
    if year in year_counts:
        year_counts[year]+=1
    else:
        year_counts[year]=1

print(year_counts)
```

```python
import datetime
dates=[datetime.datetime(year=int(row[1]),month=int(row[2]),day=1) for row in data]
print(dates[:5])

date_counts={}
for date in dates:
    if date in date_counts:
        date_counts[date]+=1
    else:
        date_counts[date]=1
print(date_counts)
```

```python
sex_counts={}
sex=[row[5] for row in data]
for item in sex:
    if item in sex_counts:
        sex_counts[item]+=1
    else:
        sex_counts[item]=1

race_counts={}
race=[row[7] for row in data]
for item in race:
    if item in race_counts:
        race_counts[item]+=1
    else:
        race_counts[item]=1
```

```python
print(sex_counts)
print(race_counts)
```

```python
with open("census.csv","r") as f:
    census=list(csv.reader(f))
    print(census)
```

```python
pop={}
pop["White"]=census[1][10]
pop["Hispanic"]=census[1][11]
pop["Black"]=census[1][12]
pop["Native American/Native Alaskan"]=census[1][13]
pop["Asian/Pacific Islander"]=census[1][14]+census[1][15]

race_per_hundred={}
for race in race_counts:
    race_per_hundred[race]=int(race_counts[race])/int(pop[race])*100000

print(race_per_hundred)

                       
                       
```
